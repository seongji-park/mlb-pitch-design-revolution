```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pybaseball import statcast

plt.style.use('seaborn-v0_8-whitegrid')
print("Setup complete!")
```

```python
seasons = {}
sample_periods = [
    (2016, '2016-04-03', '2016-04-17'),
    (2018, '2018-03-29', '2018-04-12'),
    (2020, '2020-07-23', '2020-08-06'),
    (2022, '2022-04-07', '2022-04-21'),
    (2024, '2024-03-28', '2024-04-11'),
    (2025, '2025-03-27', '2025-04-10'),
    (2026, '2026-03-28', '2026-04-11'),
]

for year, start, end in sample_periods:
    print(f"Fetching {year}...")
    data = statcast(start_dt=start, end_dt=end)
    seasons[year] = data
    print(f"  {year}: {len(data)} pitches")

print("\nAll seasons loaded!")
```

```python
pitch_evolution = {}

for year, data in seasons.items():
    mix = data['pitch_type'].value_counts(normalize=True) * 100
    pitch_evolution[year] = mix

pitch_df = pd.DataFrame(pitch_evolution).T.fillna(0)

key_pitches = ['FF', 'SL', 'CH', 'CU', 'SI', 'FC', 'ST', 'KC', 'FS']
available = [p for p in key_pitches if p in pitch_df.columns]

print("=== Pitch Type Usage % by Season ===")
print(pitch_df[available].round(1))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 7))

pitch_labels = {
    'FF': '4-Seam Fastball', 'SL': 'Slider', 'CH': 'Changeup',
    'CU': 'Curveball', 'SI': 'Sinker', 'FC': 'Cutter',
    'ST': 'Sweeper', 'KC': 'Knuckle Curve', 'FS': 'Splitter'
}

pitch_colors = {
    'FF': '#e74c3c', 'SL': '#3498db', 'CH': '#2ecc71',
    'CU': '#f39c12', 'SI': '#1abc9c', 'FC': '#95a5a6',
    'ST': '#9b59b6', 'KC': '#e67e22', 'FS': '#34495e'
}

# Left: All pitch types over time
for pitch in available:
    values = [pitch_df.loc[y, pitch] if y in pitch_df.index else 0 
              for y in sorted(seasons.keys())]
    if max(values) > 2:
        ax1.plot(sorted(seasons.keys()), values, 
                 linewidth=2.5, marker='o', markersize=6,
                 color=pitch_colors.get(pitch, '#95a5a6'),
                 label=pitch_labels.get(pitch, pitch))

ax1.set_xlabel('Season', fontsize=12, fontweight='bold')
ax1.set_ylabel('Usage %', fontsize=12, fontweight='bold')
ax1.set_title('MLB Pitch Type Usage Trends\n(Opening 2-week sample per season)',
              fontsize=14, fontweight='bold')
ax1.legend(fontsize=9, loc='upper right')
ax1.grid(True, alpha=0.3)

# Right: Sweeper emergence (if data exists)
if 'ST' in pitch_df.columns:
    sweeper_data = pitch_df['ST']
    ax2.bar(sweeper_data.index, sweeper_data.values,
            color='#9b59b6', edgecolor='white', linewidth=2)
    for x, y in zip(sweeper_data.index, sweeper_data.values):
        if y > 0.1:
            ax2.text(x, y + 0.3, f'{y:.1f}%', ha='center', 
                     fontsize=11, fontweight='bold')
    ax2.set_xlabel('Season', fontsize=12, fontweight='bold')
    ax2.set_ylabel('Usage %', fontsize=12, fontweight='bold')
    ax2.set_title('The Sweeper Explosion\nFrom non-existent to essential',
                  fontsize=14, fontweight='bold')
else:
    ax2.text(0.5, 0.5, 'Sweeper data not available\nin selected seasons',
             transform=ax2.transAxes, ha='center', fontsize=14)
    ax2.set_title('Sweeper Usage', fontsize=14, fontweight='bold')

plt.suptitle("The Pitch Design Revolution — How MLB Pitching Evolved (2016-2026)",
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('../visualizations/01_pitch_type_trends.png', dpi=150, bbox_inches='tight')
print("Chart saved!")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14, 7))

avg_velo_by_year = {}
for year, data in seasons.items():
    fb_data = data[data['pitch_type'].isin(['FF', 'SI'])]['release_speed'].dropna()
    if len(fb_data) > 0:
        avg_velo_by_year[year] = fb_data.mean()

years = sorted(avg_velo_by_year.keys())
velocities = [avg_velo_by_year[y] for y in years]

ax.plot(years, velocities, color='#e74c3c', linewidth=3, 
        marker='o', markersize=10)

for yr, velo in zip(years, velocities):
    ax.annotate(f'{velo:.1f}', xy=(yr, velo),
                xytext=(0, 12), textcoords='offset points',
                ha='center', fontsize=11, fontweight='bold')

ax.set_xlabel('Season', fontsize=13, fontweight='bold')
ax.set_ylabel('Avg Fastball Velocity (mph)', fontsize=13, fontweight='bold')
ax.set_title("MLB Average Fastball Velocity (2016-2026)\n"
             "The velocity arms race continues",
             fontsize=16, fontweight='bold', pad=15)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('../visualizations/02_velocity_inflation.png', dpi=150, bbox_inches='tight')
print("Chart saved!")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14, 7))

unique_pitches_by_year = {}
for year, data in seasons.items():
    pitcher_arsenal = data.groupby('pitcher')['pitch_type'].nunique()
    unique_pitches_by_year[year] = pitcher_arsenal.mean()

years = sorted(unique_pitches_by_year.keys())
avg_pitches = [unique_pitches_by_year[y] for y in years]

ax.bar(years, avg_pitches, color='#3498db', edgecolor='white', 
       linewidth=2, width=1.2)

for yr, val in zip(years, avg_pitches):
    ax.text(yr, val + 0.05, f'{val:.2f}', ha='center', 
            fontsize=11, fontweight='bold')

ax.set_xlabel('Season', fontsize=13, fontweight='bold')
ax.set_ylabel('Avg Pitch Types per Pitcher', fontsize=13, fontweight='bold')
ax.set_title("Arsenal Diversification — Pitchers Are Throwing More Pitch Types\n"
             "The days of fastball-slider-only are numbered",
             fontsize=15, fontweight='bold', pad=15)
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('../visualizations/03_arsenal_diversification.png', dpi=150, bbox_inches='tight')
print("Chart saved!")
plt.show()
```
